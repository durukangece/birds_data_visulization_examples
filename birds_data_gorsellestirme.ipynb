```python
birds.info() # dosya hakkında genel bilgileri görebilmek için 
```

```python
sns.set(style="darkgrid")
sns.barplot(
    x="Genus",
    y="MaxLength",        
    data=birds,
    color="cyan"
)
plt.figure(figsize=(18,12))
```

# verileri Dendrocygna/Anser şeklinde ayıralım: 

```python
g = sns.FacetGrid(birds, col="Category", height=4, aspect=.5)
g.map(sns.barplot, "Genus", "MinBodyMass", order=["MaxLength", "MinLength"])
```

```python
total = birds.groupby('Category')['Genus'].sum().reset_index()
```

```python
bar1 = sns.barplot(x="Category",  y="Genus", data=total, color='darkblue')
```

```python
Family = birds[birds.Family=='Anatidae']
```

```python
bar2 = sns.barplot(
    x='Category',
    y='Genus',
    data=Family,
    estimator=sum,
    errorbar=None,
    color='lightblue'
)    
```

```python
bar1 = sns.barplot(
    x='Category', 
    y="Genus", 
    data=Family, 
    color='darkblue')

bar2 = sns.barplot(
    x='Category', 
    y="Genus", 
    data= Family,
    estimator=sum, 
    errorbar=None,
    color='red')
```

```python
sns.histplot(data=birds, x='MaxLength', kde=True)
```

```python
sns.regplot(x=birds['MaxLength'], y=birds['MinLength'])
```

```python
sns.regplot(
    x=birds['MaxLength'], 
    y=birds['MinLength'], 
    line_kws={
        "color":"red",
        "alpha":0.6,
        "lw":5
    
    })
```

noktaların şeklini değiştirmek için:

```python
sns.regplot(
    x=birds['MaxLength'], 
    y=birds['MinLength'], 
    marker='+',
    line_kws={
        "color":"red",
        "alpha":0.6,
        "lw":5
    
    })
```

```python
sns.regplot(
    x=birds['MaxLength'], 
    y=birds['MinLength'], 
    marker='+',
    line_kws={
        "color":"red",
        "alpha":0.6,
        "lw":5
    },
    scatter_kws={
        "color":"darkblue",
        "s":50
    }
    )
```

```python
sns.lmplot(
    x='MaxLength', 
    y='MinLength', 
    data=birds,
    hue="Family",
    )
```

```python
sns.boxplot(x=birds['Family'], y=birds['MinLength'])
```

```python
birds.boxplot(column='MinLength', by='Family')
plt.show()
```
